### Gini Impurity

$$gini = 2p(1-p) \in [0, 0.5]$$

0: pure (same class)

0.5: impure


```
def gini_impurity(yes,no):
  total = yes + no
  p = yes / total
  gini = 2 * p * (1-p)
  return gini

print("A: Left: ", gini_impurity(yes = 10, no = 40))
print("A: Right: ", gini_impurity(yes = 40, no = 10))
print("B: Left: ", gini_impurity(yes = 5, no = 5))
print("B: Right: ", gini_impurity(yes = 45, no = 45))

p = np.arange(0,1,0.01)
gini = 2 * p * (1-p)
plt.plot(p,gini)
```

### Entropy

$$ entropy = -[p\log_2 p + (1-p)\log_2(1-p)] \in [0,1] $$

0: pure (same class)

1: impure



```
def entropy_function(p):
  entropy = -(p * np.log2(p) + (1-p) * np.log2(1-p) )
  return entropy

p = np.arange(0.01,1,0.01)
plt.plot(p,entropy_function(p))
```

### Information Gain
$$ information\_gain = H(S) - \frac{|A|}{|S|}H(A)- \frac{|B|}{|S|}H(B) $$
where H is the impurity measure (gini or entropy);
S: original dataset, A: first group, B: second group, |A|: size of A



```
S = 50 + 50
H_S = 0.5
A = 25
H_A = 0.4
B = 75
H_B = 0.2

InfoGain = H_S - A/S * H_A - B/S * H_B  
InfoGain
```
